```python
import pandas as pd
```

```python
data_og = pd.read_excel('reto.xlsx', sheet_name='protocolos_produccion', skiprows = 1)
data_og.head()
```

```python
data = data_og.copy() 
```

```python
# Seleccion de datos numericos
data.drop(columns=['Fecha', 'Hora', 'Tipo'], inplace = True)
```

```python
# Calculo de la media de los valores por lote
df2 = data.groupby(['indice_lote', 'Estación']).mean()
df2.head()
```

```python
# Importación de df friabilidad
df_friab = pd.read_excel('reto.xlsx', sheet_name='calidad_friabilidad')
df_friab.head()
```

```python
# Merge df base y df friabilidad
df = df2.merge(df_friab, how = 'left', on = 'indice_lote')
df
```
